```python
def get_marital_status(row):
    if row['marital_Divorced'] == 1:
        return 'Divorced'
    elif row['marital_Married'] == 1:
        return 'Married'
    elif row['marital_Single'] == 1:
        return 'Single'
    elif row['marital_Together'] == 1:
        return 'Together'
    elif row['marital_Widow'] == 1:
        return 'Widow'
    else:
        return 'Unknown'
df['Marital'] = df.apply(get_marital_status, axis=1)
df
```

```python
def get_Education(row):
    if row['education_2n Cycle'] == 1:
        return 2
    elif row['education_Basic'] == 1:
        return 1
    elif row['education_Graduation'] == 1:
        return 3
    elif row['education_Master'] == 1:
        return 4
    elif row['education_PhD'] == 1:
        return 5
    else:
        return 'Unknown'
df['Education_status'] = df.apply(get_Education,axis=1)
df.head()
```

```python
def get_relationship(row):
    if row['marital_Married'] ==1:
        return 1
    elif row['marital_Together'] == 1:
        return 1
    else:
        return 0
df['In_relationship'] = df.apply(get_relationship, axis=1)
df.head() 
```

```python
#clustering
from sklearn.cluster import KMeans
```

```python
#standardisation (Down to unit variance)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
cols_for_clustering = ['Income', 'MntTotal', 'In_relationship']
df_scaled = df.copy()
df_scaled[cols_for_clustering] = scaler.fit_transform(df[cols_for_clustering])
df_scaled[cols_for_clustering].describe()
```

```python
#choosing range for cluster values for elbow method.
X = df_scaled[cols_for_clustering]
inertia_list = []
for K in range(2,10):
    inertia = KMeans(n_clusters=K, random_state=7).fit(X).inertia_
    inertia_list.append(inertia)
```

```python
plt.figure(figsize=[7,5])
plt.plot(range(2,10), inertia_list, color=(54 / 255, 113 / 255, 130 / 255))
plt.title("Inertia vs. Number of Clusters")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.show()
```

```python
#fitting desired numbeer of clusters into our model and creating segments.
model = KMeans(n_clusters=4, random_state = 7)
model.fit(df_scaled[cols_for_clustering])
df_scaled['Cluster'] = model.predict(df_scaled[cols_for_clustering])

```

```python
#Generating 2 most relevant components/properties of clusters.
from sklearn import decomposition
pca = decomposition.PCA(n_components = 2)
pca_res = pca.fit_transform(df_scaled[cols_for_clustering])
df_scaled['pc1'] = pca_res[:,0]
df_scaled['pc2'] = pca_res[:,1]
```

```python
#Visualisation 
plt.figure(figsize=(8, 6))
sns.scatterplot(x='pc1', y='pc2', data=df_scaled, hue='Cluster', palette='viridis')
plt.title('Clustered Data Visualization')
plt.xlabel('Principal Component 1 (pc1)')
plt.ylabel('Principal Component 2 (pc2)')
plt.legend(title='Clusters')
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Cluster', y='In_relationship', data=df, palette='RdPu')
plt.title('In_relationship by cluster')
plt.xlabel('Cluster')
plt.ylabel('In_relationship')
```

```python
df['Cluster'] = df_scaled.Cluster
df.groupby('Cluster')[cols_for_clustering].mean()
```

```python
#Number of customers per cluster
customer_num = df_scaled.groupby('Cluster')['MntTotal'].count().reset_index()

#vizualisation
plt.figure(figsize=(8,6))
sns.barplot(x='Cluster', y='MntTotal', data=customer_num, palette = 'RdPu')
plt.title('Cluster sizes')
plt.xlabel('Cluster')
plt.ylabel('MntTotal')
```

```python
#avg income by cluster
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Income', y='MntTotal', data=df, hue = 'Cluster', palette='RdPu')
plt.title('Income by cluster')
plt.xlabel('Income')
plt.ylabel('MntTotal')
plt.legend(title='Clusters')
```

```python
#forming clusters without modifying the original values of columns used to form clusters
scaler = StandardScaler()
col_to_cluster = ['Income','MntTotal','In_relationship','Education_status']
df_clustered = df.copy()

model = KMeans(n_clusters=4, random_state= 7)
model.fit(df_clustered[col_to_cluster])
df_clustered['cluster'] = model.predict(df_clustered[col_to_cluster])
df_clustered.head()
```

```python
#average income of each cluster
df_clustered['Avg_income'] = df_clustered.groupby('Cluster')['Income'].mean()

#barplot
plt.figure(figsize=(8,6))
sns.barplot(x='Cluster', y='Avg_income',data=df_clustered,palette="viridis")
plt.title('cluster Avg income')
plt.xlabel('Cluster')
plt.ylabel('Avg_income')
plt.show()
```
